```r
d=c(
  845.00, 844.00, 840.00, 840.00,

  838.00, 837.00, 841.00, 841.00,

  850.00, 850.00, 835.00, 835.00,

  844.00, 843.00, 844.00, 845.00,

  846.00, 847.00, 846.00, 845.00)
```

```r
install.packages("moments")
library(moments)
```

```r
shapiro.test(d)
jarque.test(d)
anscombe.test(d)
```

q2.png

```r
#Mean interval estimation with Unkown variance
n=length(d)
alpha=0.05
sd_d=sd(d)
mean_d=mean(d)
t=qt(alpha/2,n-1)

cat("n=",n,"\n",
    "sd=",sd_d,"\n",
    "mean=",mean_d,"\n")

lower_mean = mean_d - (abs(t) * sd_d / sqrt(n))
upper_mean = mean_d + (abs(t) * sd_d / sqrt(n))

cat("[",lower_mean,",",upper_mean,"]")
```

q3.png

```r
#Standard deviation interval estimation

var_d=(sd_d)^2

lower_sd=sqrt((var_d)*(n-1)/qchisq(1-alpha/2,n-1))
upper_sd=sqrt((var_d)*(n-1)/qchisq(alpha/2,n-1))

cat("[",lower_sd,",",upper_sd,"]")
```

q4.png

```r
P830 = pnorm(830, mean_d, sd_d)
P860 = pnorm(860, mean_d, sd_d, lower.tail = FALSE)

cat(P830*10^6+P860*10^6)
```

```r
#proportion interval estimation

p=0.09
n2=100
alpha=0.05

lower_pi = p - qnorm(alpha/2)*sqrt(p*(1-p)/n2)
upper_pi = p + qnorm(alpha/2)*sqrt(p*(1-p)/n2)

cat("[",lower_pi*100,",",upper_pi*100,"]")
```
