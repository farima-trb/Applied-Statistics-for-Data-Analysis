```r
#Create a population with uniform Dist

set.seed(200)
pop=runif(10000,min=0,max=1)
hist(pop,main = "Dist of Population")

mean_pop=(0+1)/2
mean_pop
var_pop=(1-0)^2/12
#in central limit theorem formula --> var_dist = var/n
var_pop/30
```

```r
#Getting 500 samples with size=30
sample_dist=data.frame(matrix(nrow = 500,ncol = 30))
for (i in 1:500){
  sample_dist[i,]=sample(pop,30)
}

head(sample_dist)
```

```r
sample_dist[,"mean"]=rowMeans(sample_dist[,c(1:30)])
head(sample_dist)
```

```r
#Dist of Sample Means
hist(sample_dist$mean,breaks = 20,main = "Dist of Sample Means")
mean(sample_dist$mean)
var(sample_dist$mean)
```
