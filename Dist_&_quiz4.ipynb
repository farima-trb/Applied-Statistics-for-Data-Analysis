```r
#Binomial Distribution

p=20/100
n=12

#probablity of no sales 
cat("P(X=0) =",dbinom(0,n,p),"\n")

#probablity of exactly 2 sales
cat("P(X=2) =",dbinom(2,n,p),"\n")

#probablity of at most 2 sales
cat("P(X<=2) =",pbinom(2,n,p,lower.tail = T),"\n")

#probablity of at least 2 sales
cat("P(X=>2) =",pbinom(1,n,p,lower.tail = F),"\n")
```

```r
#Hypergeometric Distribution

N=200
n=30
p=0.05
r=0

#probability of no successes
cat("P(X=0) =",dhyper(r,p*N,(1-p)*N,n))
```

```r
#Geometric Distribuution

p=0.03
x=100

#probability that the first success will be the 100th part
cat("P(X=100) =",dgeom(x,p))
```

```r
#Poisson Distribution

lambda=5

#probability of 3 accidents
cat("P(X=3) =",dpois(3,lambda),"\n")

#probability of at least 5 accidents
cat("P(X>=5) =",ppois(4,lambda, lower.tail = F))
```
