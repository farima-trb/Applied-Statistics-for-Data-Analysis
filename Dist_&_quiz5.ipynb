```r
#Normal Distribution
cat("Area:",pnorm(6.2,mean=5,sd=10,lower.tail=T)-pnorm(5,mean=5,sd=10,lower.tail=T))
```

```r
#test for normality

#create sample from a normal distribution
set.seed(42)
x1=rnorm(100,1.5,1.8)
mean(x1)

#create sample from a non-normal distribution
set.seed(42)
x2=rweibull(100,1,1.5)
mean(x2)
```

```r
hist(x1,freq = F,breaks = 15)
lines(density(x1),col = "red",lwd = 2)

hist(x2,freq = F,breaks = 15)
lines(density(x2),col = "blue",lwd = 2)
```

```r
#QQ plot

qqnorm(x1,main="QQ Plot of x1")
qqline(x1,col="red")

qqnorm(x2,main="QQ Plot of x2")
qqline(x2,col="blue")
```

```r
#Shapiro test --> good for sample_size <=25
# our sample_size=100

shapiro.test(x1)
shapiro.test(x2)
```

```r
# p_value in x1 > 0.05 --> normal dist --> T
# p_value in x2 < 0.05 --> normal dist --> F
```

```r
# Skewness and Kurtosis test

install.packages("moments")
library(moments)
```

```r
#Jarque-Bera test (checks skewness=0 ?)

jarque.test(x1)
jarque.test(x2)
```

```r
# p_value in x1 > 0.05 --> normal dist --> T
# p_value in x2 < 0.05 --> normal dist --> F
```

```r
#Anscombe-Glynn test (checks kurtosis=3 ?)

anscombe.test(x1)
anscombe.test(x2)
```

```r
# p_value in x1 > 0.05 --> normal dist --> T
# p_value in x2 < 0.05 --> normal dist --> F
```

Transform to Normal Dist

```r
#sqrt(x2)
x2_trans1<-sqrt(x2)

hist(x2,freq = F,breaks = 15)
lines(density(x2),col = "blue",lwd = 2)

hist(x2_trans1,freq = F,breaks = 15)
lines(density(x2_trans1),col="darkblue",lwd = 2)

qqnorm(x2,main="QQ Plot of x2")
qqline(x2,col="blue",lwd = 2)

qqnorm(x2_trans1,main="QQ Plot of x2_trans1")
qqline(x2_trans1,col="darkblue",lwd = 2)
```

```r
shapiro.test(x2_trans1)
cat("-------------------------------------------------")
jarque.test(x2_trans1)
cat("-------------------------------------------------")
anscombe.test(x2_trans1)
```

```r
# p_value in Shapiro test < 0.05 --> normal dist --> F
# p_value in Jarque-Bera test < 0.05 --> normal dist --> F
# p_value in Anscombe-Glynn test < 0.05 --> normal dist --> F
```

```r
#log(x2)
x2_trans2<-log(x2)

hist(x2,freq = F,breaks = 15)
lines(density(x2),col = "blue",lwd = 2)

hist(x2_trans2,freq = F,breaks = 15)
lines(density(x2_trans2),col="darkblue",lwd = 2)

qqnorm(x2,main="QQ Plot of x2")
qqline(x2,col="blue",lwd = 2)

qqnorm(x2_trans2,main="QQ Plot of x2_trans1")
qqline(x2_trans2,col="darkblue",lwd = 2)
```

```r
shapiro.test(x2_trans2)
cat("-------------------------------------------------")
jarque.test(x2_trans2)
cat("-------------------------------------------------")
anscombe.test(x2_trans2)
```

```r
# p_value in Shapiro test > 0.05 --> normal dist --> T
# p_value in Jarque-Bera test > 0.05 --> normal dist --> T
# p_value in Anscombe-Glynn test > 0.05 --> normal dist --> T
```

```r
#Exponential Distribution

cat("P(X=50) =",pexp(50,rate=1/100,lower.tail = F))
```

```r
#test for normality

#create sample from a exponential distribution
set.seed(42)
x1=rexp(100)
mean(x1)

#create sample from a non-exponential distribution
set.seed(42)
x2=rchisq(100,1)
mean(x2)
```

```r
hist(x1,freq=F,breaks=15)
lines(density(x1),col = "red",lwd = 2)
hist(x2,freq=F,breaks=15)
lines(density(x1),col = "blue",lwd = 2)
```

```r
install.packages("exptest")
library("exptest")
```

```r
shapiro.exp.test(x1)

shapiro.exp.test(x2)
```

```r
# p_value in x1 < 0.05 --> normal dist --> F
# p_value in x2 < 0.05 --> normal dist --> F
```
